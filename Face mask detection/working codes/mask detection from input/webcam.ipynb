```python
from google.colab import drive
drive.mount('/content/drive')

from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from keras.models import load_model
import numpy as np

#importing previously saved model to current work
model_path = '/content/drive/MyDrive/Assignment asset/saved files/model.hdf5'
model = load_model(model_path)

#print(model.summary())

#capturing video frames from webcam usign OpenCV
import cv2

video_capture = cv2.VideoCapture(0)

while True:
  ret, frame = video_capture.read()         #for capturing frame by frame

  #matching up the frames with the input shape of the imported model

  #checking if the frame is empty
  if not ret:
    print("Error: Failed to capture frame from webcam")
    break

  #resizing the frame
  if frame.shape[0] != 0 and frame.shape[1] != 0:
    frame = cv2.resize(frame, (224, 224))
  else:
    print("Error: Frame shape is empty")
    break

  #converting the color space to RGB
  frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

  #converting the frame to numpy array
  image = np.array(frame)

  #preprocessing the image
  image = preprocess_input(image)

  #adding extra dimension to the image to represent the batch size 1
  image = np.expand_dims(image, axis = 0)

  preprocessed_frame = image

  #making prediction using imported model
  prediction = model.predict(preprocessed_frame)

  #displaying the prediction using the model
  if prediction [0][0] > 0.5:
    label = 'Without Mask'
    color = (0, 0, 255)
  else:
    label = 'With Mask'
    color = (0, 255, 0)

  cv2.putText(frame, label, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)

  #displaying the resulting frame
  cv2.imshow('Video', frame)

  #breaking the loop upon pressing 'x'
  if cv2.waitKey(1) & 0xFF == ord('x'):
    break

#releasing the capture
video_capture.release()
cv2.destroyAllWindows()
  
```
